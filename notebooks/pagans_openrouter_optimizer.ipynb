# OpenRouter Optimizer Notebook

Set `OPENROUTER_API_KEY` before running.


```python
from pagans import PAGANSOptimizer, Provider
```

```python
async def main():
    async with PAGANSOptimizer(provider=Provider.OPENROUTER) as optimizer:
        result = await optimizer.optimize(
            prompt='Refactor this algorithm for readability and maintainability.',
            target_model='claude-sonnet-4',
        )
        print(result.optimized)

await main()
```
