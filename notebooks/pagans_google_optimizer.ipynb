# Google AI Studio Optimizer Notebook

Set `GOOGLE_API_KEY` before running.


```python
from pagans import PAGANSOptimizer, Provider
```

```python
async def main():
    async with PAGANSOptimizer(
        provider=Provider.GOOGLE,
        optimizer_model='gemini-2.5-flash',
    ) as optimizer:
        result = await optimizer.optimize(
            prompt='Create a concise onboarding checklist for backend engineers.',
            target_model='gemini-2.5-pro',
        )
        print(result.optimized)

await main()
```
