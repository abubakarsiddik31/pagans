# Anthropic Optimizer Notebook

Set `ANTHROPIC_API_KEY` before running.


```python
from pagans import PAGANSOptimizer, Provider
```

```python
async def main():
    async with PAGANSOptimizer(
        provider=Provider.ANTHROPIC,
        optimizer_model='claude-sonnet-4-20250514',
    ) as optimizer:
        result = await optimizer.optimize(
            prompt='Turn this feature request into a precise technical specification.',
            target_model='claude-sonnet-4',
        )
        print(result.optimized)

await main()
```
