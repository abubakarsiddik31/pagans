# Z.ai Optimizer Notebook

Set `ZAI_API_KEY` before running.


```python
from pagans import PAGANSOptimizer, Provider
```

```python
async def main():
    async with PAGANSOptimizer(
        provider=Provider.ZAI,
        optimizer_model='glm-4.7',
    ) as optimizer:
        result = await optimizer.optimize(
            prompt='Rewrite this operational runbook to be easier to execute on-call.',
            target_model='gpt-4o',
        )
        print(result.optimized)

await main()
```
