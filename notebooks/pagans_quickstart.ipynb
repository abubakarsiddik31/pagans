# PAGANS Quickstart

This notebook shows single optimization, comparison, and batch usage.


```python
from pagans import PAGANSOptimizer, Provider
```

```python
async def run_single():
    async with PAGANSOptimizer(provider=Provider.OPENROUTER) as optimizer:
        result = await optimizer.optimize(
            prompt='Design a robust retry strategy for external APIs.',
            target_model='gpt-4o',
        )
        print(result.optimized)

await run_single()
```

```python
async def run_compare_and_batch():
    async with PAGANSOptimizer(provider=Provider.OPENROUTER) as optimizer:
        compare = await optimizer.compare_optimizations(
            prompt='Draft a clean architecture guideline.',
            target_models=['gpt-4o', 'claude-sonnet-4', 'gemini-2.5-pro'],
        )
        print('Compare models:', list(compare.keys()))

        batch = await optimizer.optimize_multiple(
            prompts=['Prompt A', 'Prompt B'],
            target_model='gpt-4o',
        )
        print('Batch count:', len(batch))

await run_compare_and_batch()
```
