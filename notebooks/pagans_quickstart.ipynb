# PAGANS Quickstart Notebook

This notebook shows how to optimize prompts for different model families using PAGANS with OpenRouter.

````python
import asyncio
from pagans import PAGANSOptimizer
````

## 2) Single Prompt Optimization


````python
async def single_optimization():
    async with PAGANSOptimizer() as optimizer:
        result = await optimizer.optimize(
            prompt="Write a robust Python function that retries HTTP requests with exponential backoff.",
            target_model="gpt-5.4"
        )

        print("Target model:", result.target_model)
        print("Target family:", result.target_family.value)
        print("Optimization time:", f"{result.optimization_time:.2f}s")
        print("\nOptimized prompt:\n")
        print(result.optimized)

await single_optimization()
````

## 3) Compare Families (OpenAI, Anthropic, xAI, Gemini)


````python
async def compare_families():
    prompt = "Design a production-ready API error handling strategy."
    target_models = [
        "gpt-5.4",
        "claude-sonnet-4.6",
        "grok-4.20-beta",
        "gemini-3.1-pro-preview",
    ]

    async with PAGANSOptimizer() as optimizer:
        results = await optimizer.compare_optimizations(
            prompt=prompt,
            target_models=target_models,
        )

    for model, result in results.items():
        if hasattr(result, "target_family"):
            print(f"{model} -> {result.target_family.value}")
        else:
            print(f"{model} -> error: {result}")

await compare_families()
````

## 4) Batch Optimization


````python
async def batch_optimization():
    prompts = [
        "Summarize this architecture decision record.",
        "Generate unit tests for this function.",
        "Write concise release notes from commit messages.",
    ]

    async with PAGANSOptimizer() as optimizer:
        results = await optimizer.optimize_multiple(
            prompts=prompts,
            target_model="grok-4-fast",
            max_concurrent=3,
        )

    for idx, result in enumerate(results, start=1):
        print(f"{idx}. {result.target_model} ({result.target_family.value})")

await batch_optimization()
````
