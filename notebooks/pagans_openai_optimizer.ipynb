# OpenAI Optimizer Notebook

Set `OPENAI_API_KEY` before running.


```python
from pagans import PAGANSOptimizer, Provider
```

```python
async def main():
    async with PAGANSOptimizer(
        provider=Provider.OPENAI,
        optimizer_model='gpt-4.1-mini',
    ) as optimizer:
        result = await optimizer.optimize(
            prompt='Write testing guidelines for a Python package.',
            target_model='gpt-4o',
        )
        print(result.optimized)

await main()
```
